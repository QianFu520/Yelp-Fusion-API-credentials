```python
import json
with open('/Users/qfu88/.secret/yelp_api.json') as f: 
    login = json.load(f)
login.keys()
```

```python
# import the YelpAPI Class
from yelpapi import YelpAPI
# Create an instance with your key
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
search_results = yelp_api.search_query(location='Chicago, IL',
                                       term='Sandwich')
print(type(search_results))
search_results.keys()
```

```python
search_results['total']
```

```python
import pandas as pd
Chicago = pd.DataFrame(search_results['businesses'])
Chicago.head(2) 
```

```python
search_results20 = yelp_api.search_query(location='Chicago, IL',
                                       term='Sandwich', offset = 20)
print(type(search_results20))
search_results20.keys()
```

```python
Chicago20 = pd.DataFrame(search_results20['businesses'])
Chicago20.head(2)
```

```python
Chicago_businesses = pd.concat([Chicago, Chicago20],
                      ignore_index=True)
display(Chicago_businesses.head(3), Chicago_businesses.tail(3))
```

```python
Chicago_businesses.to_json(orient='records')
```
